```python
Q1. Write a python program to extract the video URL of the first five videos.

To extract the video URLs of the first five videos from the YouTube channel "PW-Foundation," we need to use the YouTube Data API rather than scraping the web page directly. The YouTube Data API allows us to retrieve information about videos from YouTube channels programmatically.

Here's a Python program using the `google-api-python-client` library to fetch the video IDs of the first five videos from the specified YouTube channel and then construct the video URLs:

from googleapiclient.discovery import build

# YouTube Data API key (Replace 'YOUR_API_KEY' with your actual API key)
API_KEY = 'YOUR_API_KEY'

# YouTube channel ID of PW-Foundation
CHANNEL_ID = 'UC5NT2b3oZmRwrFxZ6VX2D1g'

# Function to fetch video URLs from the channel
def get_video_urls(api_key, channel_id, max_results=5):
    # Build the YouTube Data API service
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Call the channels.list method to retrieve channel details
    channel_response = youtube.channels().list(
        part='contentDetails',
        id=channel_id
    ).execute()

    # Extract uploads playlist ID from channel details
    uploads_playlist_id = channel_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

    # Call the playlistItems.list method to retrieve videos from the uploads playlist
    playlist_response = youtube.playlistItems().list(
        part='snippet',
        playlistId=uploads_playlist_id,
        maxResults=max_results
    ).execute()

    # Extract video IDs from playlist response
    video_ids = [item['snippet']['resourceId']['videoId'] for item in playlist_response['items']]

    # Construct video URLs
    video_urls = ['https://www.youtube.com/watch?v=' + video_id for video_id in video_ids]

    return video_urls

# Fetch video URLs of the first five videos from the channel
video_urls = get_video_urls(API_KEY, CHANNEL_ID)

# Print the video URLs
print("Video URLs of the first five videos:")
for i, url in enumerate(video_urls, start=1):
    print(f"Video {i}: {url}")

Before running this code, we need to:

1. Obtain an API key for the YouTube Data API. You can obtain one by creating a project in the Google Cloud Console and enabling the YouTube Data API.
2. Replace `'YOUR_API_KEY'` with your actual API key.
3. Replace `'UC5NT2b3oZmRwrFxZ6VX2D1g'` with the channel ID of PW-Foundation or the desired channel.
4. Install the `google-api-python-client` library if you haven't already by running `pip install google-api-python-client`.

This program fetches the video URLs of the first five videos from the specified YouTube channel using the YouTube Data API and prints them.
```

```python
Q4. Write a python program to extract the number of views of the first five videos.

To extract the number of views of the first five videos from the YouTube channel "PW-Foundation," we can once again utilize the YouTube Data API to retrieve information about the videos, including their view counts.

Here's a Python program using the `google-api-python-client` library to fetch the number of views of the first five videos from the specified YouTube channel:
from googleapiclient.discovery import build

# YouTube Data API key (Replace 'YOUR_API_KEY' with your actual API key)
API_KEY = 'YOUR_API_KEY'

# YouTube channel ID of PW-Foundation
CHANNEL_ID = 'UC5NT2b3oZmRwrFxZ6VX2D1g'

# Function to fetch video view counts from the channel
def get_video_view_counts(api_key, channel_id, max_results=5):
    # Build the YouTube Data API service
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Call the channels.list method to retrieve channel details
    channel_response = youtube.channels().list(
        part='contentDetails',
        id=channel_id
    ).execute()

    # Extract uploads playlist ID from channel details
    uploads_playlist_id = channel_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

    # Call the playlistItems.list method to retrieve videos from the uploads playlist
    playlist_response = youtube.playlistItems().list(
        part='contentDetails',
        playlistId=uploads_playlist_id,
        maxResults=max_results
    ).execute()

    # Extract video IDs from playlist response
    video_ids = [item['contentDetails']['videoId'] for item in playlist_response['items']]

    # Fetch video statistics (including view count) for each video
    videos_response = youtube.videos().list(
        part='statistics',
        id=','.join(video_ids)
    ).execute()

    # Extract view counts for each video
    view_counts = [int(item['statistics']['viewCount']) for item in videos_response['items']]

    return view_counts

# Fetch view counts of the first five videos from the channel
view_counts = get_video_view_counts(API_KEY, CHANNEL_ID)

# Print the view counts
print("Number of views of the first five videos:")
for i, count in enumerate(view_counts, start=1):
    print(f"Video {i}: {count}")
```

```python
Q5. Write a python program to extract the time of posting of video for the first five videos.

To extract the time of posting of the first five videos from the YouTube channel "PW-Foundation," we can utilize the YouTube Data API to retrieve information about the videos, including their publish dates.

Here's a Python program using the `google-api-python-client` library to fetch the publish dates of the first five videos from the specified YouTube channel:

from googleapiclient.discovery import build
from dateutil import parser

# YouTube Data API key (Replace 'YOUR_API_KEY' with your actual API key)
API_KEY = 'YOUR_API_KEY'

# YouTube channel ID of PW-Foundation
CHANNEL_ID = 'UC5NT2b3oZmRwrFxZ6VX2D1g'

# Function to fetch publish dates of videos from the channel
def get_video_publish_dates(api_key, channel_id, max_results=5):
    # Build the YouTube Data API service
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Call the channels.list method to retrieve channel details
    channel_response = youtube.channels().list(
        part='contentDetails',
        id=channel_id
    ).execute()

    # Extract uploads playlist ID from channel details
    uploads_playlist_id = channel_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

    # Call the playlistItems.list method to retrieve videos from the uploads playlist
    playlist_response = youtube.playlistItems().list(
        part='snippet',
        playlistId=uploads_playlist_id,
        maxResults=max_results
    ).execute()

    # Extract video IDs from playlist response
    video_ids = [item['snippet']['resourceId']['videoId'] for item in playlist_response['items']]

    # Fetch video details (including publish date) for each video
    videos_response = youtube.videos().list(
        part='snippet',
        id=','.join(video_ids)
    ).execute()

    # Extract publish dates for each video
    publish_dates = [parser.parse(item['snippet']['publishedAt']) for item in videos_response['items']]

    return publish_dates

# Fetch publish dates of the first five videos from the channel
publish_dates = get_video_publish_dates(API_KEY, CHANNEL_ID)

# Print the publish dates
print("Publish dates of the first five videos:")
for i, date in enumerate(publish_dates, start=1):
    print(f"Video {i}: {date}")
```
